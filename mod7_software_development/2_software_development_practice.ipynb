## Type Hinting

In this week's lecture, we talked about **type hinting**, which is a type of documentation we can add to explain data types in our code. 

In type hinting syntax, we can include an "or" operator. This tells us that a function will return one type **or** another type. 

Below is an example of a function, `plant_growth_checker`, which checks if a plant is dead or alive, and then returns the height of a live plant, or returns a message that the plant has died. Notice the type hinting in the function defintion: we will either get a `float` or a `str` from this function.

````python
def plant_growth_checker(plant_height: float, plant_alive: bool) -> float | str:
    if plant_alive:
        return plant_height
    else:
        return "This plant has died :("
    
print(type(plant_growth_checker(2.5, True)))
print(type(plant_growth_checker(3.1, False)))
````

In Python, there is a class of methods to aid in testing: the `unittest` module is a built-in tool that helps you check if your code is working correctly by letting you write and run tests. However, we can achieve the same goal using **assert** statements to verify that our functions and methods return the expected values. 

## The `assert` Statement

The assert statement is a built-in statement in Python used to, as the name says, assert if a given condition is true or not.

If the condition is true, nothing happens, but if it's not true, an error is raised.

This example returns true, thus it does not output or return anything.

```python
assert 1 > 0
```

This example is false, thus it will return an `AssertionError`.

```python
>>> assert 1 < 0

Traceback (most recent call last):
  File "<stdin>", line 1, in <module>
AssertionError
```

The basic syntax for using `assert` can be broken down as such:
```python
assert <condition being tested>, <error message to be displayed>
```

## Putting it all together

Now that we've seen how we can use type hinting and testing to help us code, let's try making our own small class with these tools.

Here is the description of our class, **Flower**:

- Attributes:

    - `type` (e.g., "Rose", "Tulip")

    - `age` (in days, as an integer)

- Methods:

    - `water()` – prints a message like "Watered the flower."

    - `is_fresh()` – returns True if age < 7

    - `describe()` – returns a string describing the flower, e.g., "This is a 3-day-old Rose."
    

Once you have your first draft of your class, write at least one test for each method and ensure that your class is working as you intended.

# Q.4: Write the `Flower` class with testing

````python
# Flower class:
class Flower:
    """Class representing a flower."""
    
    def __init__(self, flower_type: str, age: int):
        """Initialize flower with type and age in days."""
        self.type = flower_type
        self.age = age
    
    def water(self) -> str:
        """Water the flower and return a message."""
        return f"Watered the {self.type}."
    
    def is_fresh(self) -> bool:
        """Return True if flower is less than 7 days old."""
        return self.age < 7
    
    def describe(self) -> str:
        """Return a description of the flower."""
        return f"This is a {self.age}-day-old {self.type}."
    
````

````python
# Flower class testing:
# Test the Flower class
def test_flower_class():
    print("Testing Flower class...")
    
    # Create a fresh flower
    rose = Flower("Rose", 3)
    
    # Test __init__ attributes
    assert rose.type == "Rose", f"Expected 'Rose', got {rose.type}"
    assert rose.age == 3, f"Expected 3, got {rose.age}"
    print("✓ __init__ works correctly")
    
    # Test water() method
    assert rose.water() == "Watered the Rose.", f"Expected 'Watered the Rose.', got {rose.water()}"
    print("✓ water() works correctly")
    
    # Test is_fresh() method with fresh flower
    assert rose.is_fresh() == True, f"Expected True, got {rose.is_fresh()}"
    print("✓ is_fresh() returns True for fresh flower")
    
    # Test describe() method
    assert rose.describe() == "This is a 3-day-old Rose.", f"Expected 'This is a 3-day-old Rose.', got {rose.describe()}"
    print("✓ describe() works correctly")
    
    # Create an old flower
    tulip = Flower("Tulip", 10)
    
    # Test is_fresh() with old flower
    assert tulip.is_fresh() == False, f"Expected False, got {tulip.is_fresh()}"
    print("✓ is_fresh() returns False for old flower")
    
    # Test describe() for old flower
    assert tulip.describe() == "This is a 10-day-old Tulip.", f"Expected 'This is a 10-day-old Tulip.', got {tulip.describe()}"
    print("✓ describe() works correctly for old flower")
    
    # Test water() for different flower
    assert tulip.water() == "Watered the Tulip.", f"Expected 'Watered the Tulip.', got {tulip.water()}"
    print("✓ water() works correctly for different flower")
    
    print("\nAll tests passed! ✓")

# Run tests
test_flower_class()
````
